```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

train_dataset = pd.read_csv("train.csv")
train_dataset.head()
```

```python
test_dataset = pd.read_csv("test.csv")
test_dataset.head()
```

```python
len(train_dataset)
```

```python
len(test_dataset)
```

```python
# Dividindo X/y para os dados de treino
passengersIds = train_dataset["PassengerId"]
X = train_dataset.drop(["PassengerId", "Survived", "Name", "Ticket", "Cabin"], axis=1)
y = train_dataset["Survived"]
X
```

```python
# Alterando sexo para 1 ou 0
X['Sex'].replace('female', 0, inplace=True)
X['Sex'].replace('male', 1, inplace=True)
X.head()
```

```python
X.isna().sum()
```

```python
# Preenchendo idade com a média e transformando em inteiro
X["Age"].fillna(
    X["Age"].mean(),
    inplace=True
)
X["Age"] = X["Age"].astype(int)
X.isna().sum()
```

```python
# Preenchendo Embarked
X["Embarked"].fillna("S", inplace=True)
X.isna().sum()
```

```python
# Transformando categorias em números
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Pclass", "Embarked"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer(
    [("one_hot", one_hot, categorical_features)], 
    remainder="passthrough"
)
transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
passengersIds_test = test_dataset["PassengerId"]
X_test = test_dataset.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1)
X_test.head()
```

```python
# Alterando sexo para 1 ou 0
X_test['Sex'].replace('female', 0, inplace=True)
X_test['Sex'].replace('male', 1, inplace=True)
X_test.head()
```

```python
X_test.isna().sum()
```

```python
# Preenchendo idade com a média e transformando em inteiro
X_test["Age"].fillna(
    X_test["Age"].mean(),
    inplace=True
)
X_test["Age"] = X_test["Age"].astype(int)
X_test.isna().sum()
```

```python
# Preencher Fare com média
X_test["Fare"].fillna(
    X_test["Fare"].mean(),
    inplace=True
)
X_test.isna().sum()
```

```python
X_test.head()
```

```python
# Transformando categorias em números
categorical_features = ["Pclass", "Embarked"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer(
    [("one_hot", one_hot, categorical_features)], 
    remainder="passthrough"
)
transformed_X_test = transformer.fit_transform(X_test)
transformed_X_test
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(transformed_X, y)

predictions = model.predict(transformed_X_test)

X_test["Survived"] = predictions
```

```python
model.score(transformed_X, y)
```
